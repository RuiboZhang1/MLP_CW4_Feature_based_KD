```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
train_agreement_high = np.array([99.948, 93.92, 89.7739, 86.9817,84.912])
train_agreement_mid = np.array([99.944, 93.8339, 89.6543, 86.7626,84.782])
train_agreement_low = np.array([99.92, 93.6997, 89.65, 86.76,84.75])

datasize = ["25K", "31.3K", "37.6K", "43.8K", "50K"]

plt.figure(figsize=(8, 2),dpi=100)
plt.rcParams['axes.axisbelow'] = True
plt.plot(train_agreement_mid, marker='o', markeredgecolor="black", markeredgewidth=0.3, ms=9)
plt.fill_between([0,1,2,3,4], train_agreement_low*0.99, train_agreement_high*1.01, alpha=0.3)

plt.xticks([0,1,2,3,4],datasize)
plt.ylim(83, 102)
plt.yticks([100, 95, 90, 85])
plt.xlabel("Training Data Size", fontsize=12)
plt.ylabel("Train Agreement", fontsize=12)
plt.title("Fitnet Resnet44 Train Aggrement", fontsize=14)
plt.grid()
plt.savefig("Train_Agreement_with_error_2.pdf", bbox_inches = 'tight')
plt.show()
```

```python
test_agreement_high = [64.07, 66.05, 67.30, 67.34, 67.74]
test_agreement_mid = [63.45, 65.88, 67.03, 67.29, 67.72]
test_agreement_low = [63.40, 65.33, 67.11, 67.00, 67.63]

test_acc_high = [64.56, 67.07, 68.54, 70.02, 70.61] 
test_acc_mid = [64.36, 66.69, 68.52, 69.98, 70.13] 
test_acc_low = [64.08, 66.62, 68.50, 69.62, 69.87] 

test_acc_without_teacher = [63.00, 66.11, 68.24, 69.13, 71.14]

datasize = ["25K", "31.3K", "37.6K", "43.8K", "50K"]

fig, ax1 = plt.subplots(figsize=(6,2.5), dpi=100)

#ax1.rcParams['axes.axisbelow'] = True
ax1.plot(test_agreement_mid, marker='o', markeredgecolor="black", markeredgewidth=0.7, ms=9, c='royalblue', label="Teacher-Student Agreement")
ax1.fill_between([0,1,2,3,4], test_agreement_low, test_agreement_high, alpha=0.4)
ax1.set_xticks([0,1,2,3,4],datasize)
ax1.set_ylim(62, 71.5)
ax1.set_yticks([70, 68, 66, 64, 62])
ax1.set_xlabel("Training Data Size", fontsize=11)
ax1.set_ylabel("Test Agreement", fontsize=11, c='royalblue')
ax1.set_title("Fitnet Resnet44 Test Aggrement & Accuracy", fontsize=14)
ax1.tick_params(colors='royalblue', axis='y', which='both')
ax1.grid()

ax2 = ax1.twinx()
ax2.plot(test_acc_mid, color='mediumseagreen', marker='o', markeredgecolor="black", markeredgewidth=0.7, ms=9, label="Student Accuracy")
ax2.fill_between([0,1,2,3,4], test_acc_low, test_acc_high, alpha=0.4, color='limegreen')
ax2.set_ylim(62,71.5)
ax2.set_yticks([70, 68, 66, 64, 62])
ax2.set_ylabel("Test Accuracy", fontsize=11, color='mediumseagreen')
ax2.axhline(y = 63, color = 'mediumseagreen', linestyle = '--', xmin=0.045, xmax=0.955, label="Teacher Accuracy")
ax2.plot(test_acc_without_teacher, '--', label='Resnet44 Baseline Accuracy', c='red')
ax2.tick_params(colors='mediumseagreen', axis='y', which='both')

fig.legend(loc='lower center', ncol=2, bbox_to_anchor=(0.5, -0.3))
plt.savefig("test_agreement_and_test_accuracy_with_error.pdf", bbox_inches = 'tight')
plt.show()
```
